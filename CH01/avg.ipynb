```python
import numpy as np
```

```python
# 표본 평균을 구하는 코드
np.random.seed(0)  # 시드 고정
rewards = []

for n in range(1, 11):       # 10번 플레이
  reward = np.random.rand()  # 보상(무작위 수로 시뮬레이션)
  rewards.append(reward)
  Q = sum(rewards) / n
  print(Q)
```

```python
# 증분 구현으로 수정된 표본 평균을 구하는 코드
Q = 0

for n in range(1, 11):
  reward = np.random.rand()
  # Q = Q + (reward - Q) / n
  Q += (reward - Q) / n
  print(Q)
```
